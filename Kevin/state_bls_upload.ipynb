```python
# Dependencies and Setup
import pandas as pd
import pymongo

```

```python
# Define Mongo Connection
conn = 'mongodb://localhost:27017'

# Pass connection to the pymongo instance.
client = pymongo.MongoClient(conn)

# Connect to a database. Will create one if not already available.
db = client.bls_data



# File to Load (Remember to Change These)
file_to_load = "Resources/state_M2017_dl.csv"

# Read Purchasing File and store into Pandas data frame
state_BLS_df = pd.read_csv(file_to_load)

state_BLS_df.head()
```

```python
# Convert DF to json for loading into mongo
state_BLS_dict = {}
state_BLS_dict = state_BLS_df.to_dict('records')
```

```python
db.state_data.insert_many(state_BLS_dict)
```
